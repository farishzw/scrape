## urllib的使用

### 1. 发送请求--request
**urllib.request请求模块**最基本的**urlopen**方法
- 返回HTTPResponse类型，主要包含read,readinto,getheader方法，以及msg,status,closed等属性

```python
import urllib.request

response = urllib.request.urlopen('https://www.python.org')#抓取Python官网,返回HTTPResponse类型
print(response.read().decode('utf-8'))#调用read方法得到响应的网页内容
```

```python
print(response.status)#返回响应的状态码
print(response.getheaders())#返回响应的头信息
print(response.getheader('Server'))#返回头信息中的服务器
```

#### **urlopen**方法参数
- urlopen方法的API:urllib.request.urlopen(url, data=None, [timeout, ]*, cafile=None, capath=None, cadefault=False, context=None)
- data参数: 传递该参数请求方式变为**post**,该参数需要为bytes类型，需要使用bytes方法将参数转化为字节流编码格式的内容

```python
import urllib.parse
import urllib.request

data = bytes(urllib.parse.urlencode({'name':'germy'}),encoding='utf-8')
#bytes方法第一个参数为字符串（str）类型，用urllib.parse模块中的urlencode方法将字典参数转化为字符串
response = urllib.request.urlopen('https://www.httpbin.org/post',data = data)
print(response.read().decode('utf-8'))#结果中传递的参数在form字段中，说明模拟表单提交，以post方式传输数据
```

- timeout参数: 超时时间，单位为秒，请求超过超时时间没有响应，抛出异常

```python
response = urllib.request.urlopen('http://www.httpbin.org/get',timeout = 0.1)
print(response.read())
```

- **Request**: 加入Headers信息，通过Request类来构建请求
    - 构造方法：class urllib.request.Request(url, data=None, headers={}, origin_req_host=None, unverifiable=False, method=None)
    - 为什么要添加请求头：伪装浏览器，默认的User-Agent是Python-urllib,可修改该值伪装成浏览器

```python
from urllib import request,parse

url = 'http://www.httpbin.org/post'
headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36',
    'Host':'www.httpbin.org'
}
dict = {'name':'germy'}
data = bytes(parse.urlencode(dict),'utf-8')
req = request.Request(url=url,data=data,headers=headers,method='POST')
response = request.urlopen(req)
print(response.read().decode('utf-8'))
```

#### 高级请求
- Handler类:各种处理器，处理登录，Cookie，代理几乎可以实现http请求中的所有功能
    - urllib.request模块中的BaseHandler类是其他所有Handler类的父亲，提供最基本的方法default_open,protocol_request
    - HTTPDefaultErrorHandler: 处理HTTP响应错误
    - HTTPRedirectHandler: 处理重定向问题
    - HTTPCookiesHandler: 处理Cookie
    - ProxyHandler: 用于设置代理
    - HTTPPasswordMgr: 用于管理密码，维护用户名密码对照表
    - HTTPBasicAuthHandler: 管理认证
    - OpenerDirector: 深层配置，使用更底层的实例来操作，提供open方法，利用Handler类构建Open类

```python
#验证
from urllib.request import HTTPPasswordMgrWithDefaultRealm,HTTPBasicAuthHandler,build_opener
from urllib.error import URLError

username = 'admin'
password = 'admin'
url = 'https://ssr3.scrape.center/'

p = HTTPPasswordMgrWithDefaultRealm()
p.add_password(None,url,username,password)#添加用户名和密码
auth_handler = HTTPBasicAuthHandler(p)#实例化认证对象
opener = build_opener(auth_handler)#构建Opener类在发送请求时已经验证成功

try:
    result = opener.open(url)
    html = result.read().decode('utf-8')
    print(html)
except URLError as e:
    print(e.reason)
```

```python
#代理
from urllib.error import URLError
from urllib.request import ProxyHandler,build_opener

proxy_handler = ProxyHandler({
    'http':'http://127.0.0.1:7890',
    'https':'https://127.0.0.1:7890'
})#字典，键名是协议类型
opener = build_opener(proxy_handler)
try:
    response = opener.open('https://www.baidu.com')
    print(response.read().decode('utf-8'))
except URLError as e:
    print(e.reason)
```

### 2. 处理异常--error
urllib库中error模块定义了由request模块产生的异常，当问题出现时，request模块抛出error定义的异常
- URLError模块中的基类，由request模块产生的异常都可以捕获
- HTTPError为URLError的子类，专门处理HTTP请求错误
    - 三种属性：code: HTTP状态码
    - reason: 返回错误原因
    - headers: 返回请求头

```python
from urllib import request,error

try:
    response = request.urlopen('http://cuiqingcai.com/404')
except error.HTTPError as e:#先抛出子类异常
    print(e.reason,e.code,e.headers,sep='\n')
except error.URLError as e:#仔抛出父类异常
    print(e.reason)
else:
    print('Request successful')
```

### 3. 解析链接--parse
定义处理url的标准接口，实现url各部分的抽取、合并以及链接转换：file,ftp,gopher,hdl,http,https,imap等
- urlparse: 该方法实现URL的识别和分段,
    - API : urllib.parse.urlparse(urlstring,scheme='',allow_fragment=True),如果待解析的URL没有带协议可补充第二个参数，第三个参数若设置为False则fragment部门会被忽略，它会杯解析为path,params或者query的一部分，而fragment部分为空
    - 解析结果时ParseReuslt对象
    - 包含6个部分: scheme://netloc/path;params?query#fragment
        - scheme:协议
        - netloc:域名
        - path: 访问路径
        - params: 参数
        - query: 查询条件
        - fragment: 锚点用于直接定位页面内部的下拉位置
        
- urlsplit: 与urlparse方法相似，不单独解析paras会合并到path中

```python
from urllib.parse import urlparse

result = urlparse('http://www.baidu.com/index.html;user?id=5#commennt')
print(type(result))
print(result)
```

```python
from urllib.parse import urlsplit

result = urlsplit('http://www.baidu.com/index.html;user?id=5#commennt')
print(type(result))
print(result)
```

- urlunparse: urlparse的对立方法，用于**构造url**，该方法接收的参数为一个可迭代对象，长度**必须是6**
- urlunsplit: urlsplit的对立方法，与urlunparse相似，将链接各个部分组合成完整的链接，**参数长度必须为5**

```python
from urllib.parse import urlunparse

data = ['https','www.baidu.com','index.html','user','a=6','comment']
print(urlunparse(data))
```

```python
from urllib.parse import urlunsplit

data = ['https','www.baidu.com','index.html;user','a=6','comment']
print(urlunsplit(data))
```

- urljoin: 弥补urlunparse和urlunsplit方法在合并构造链接时限定参数长度的弊端，可以提供一个base_url基础链接，和新的连接作为参数
    - 一般base_url提供三项内容:scheme、netloc、和path
    - 如果新链接不存在这三项则予以补充，如果存在则base_url的信息不起作用

```python
from urllib.parse import urljoin

print(urljoin('https://baidu.com','FAQ.html'))
print(urljoin('https://baidu.com','https://cuiqingcai.com/FAQ.html'))
```

- urlencode: 可用于序列化url请求中的参数,传入参数为字典
- parse_qs: 反序列化，将一串get请求参数转回字典;parse_qsl：转化为元组组成的列表

```python
from urllib.parse import urlencode

params = {
    'name':'germey',
    'age':25
}
base_url = 'https://www.baidu.com?'
url = base_url+urlencode(params)
print(url)
```

```python
from urllib.parse import parse_qs

query = 'name=germey&age=25'
print(parse_qs(query))
```

- quote: 将内容转化为URL编码的格式，URL中带有中文参数可将中文转化为URL编码
- unquote: 将URL解码

```python
from urllib.parse import quote

keyword = '壁纸'
url='https://www.baidu.com/s?wd='+quote(keyword)
print(url)
```

```python
from urllib.parse import unquote

url='https://www.baidu.com/s?wd=%E5%A3%81%E7%BA%B8'
print(unquote(url))
```

### 1.GET请求
- GET请求获取: get方法，返回类型为requests.models.Response对象，有status_code:状态码,requests库还提供内置的状态码查询对象requests.codes；text:响应体内容,Cookie等属性
- 链接添加额外信息:get(url,params=data),以字典的形式传入url参数，避免需要构造url再请求
- 添加请求头:headers

```python
#get请求
import requests

r = requests.get('https://www.baidu.com/')#不用代理
print(type(r))
print(r.status_code)
print(type(r.text))
print(r.text[1:100])
print(r.cookies)
```

```python
#添加请求参数
import requests

data = {
    'name':'germey',
    'age':25
}
r = requests.get('https://www.httpbin.org/get',params=data)
print(r.text)
print(r.json())#调用json方法将返回结果（json格式的字符串）转化为字典
```

```python
#抓取二进制数据：图片、音频、视频
r = requests.get('https://scrape.center/favicon.ico')
print(r.content)#二进制数据
#保存调用open方法，第一个参数为文件名，第二个参数为二进制写的形式打开文件，向文件写入二进制数据
with open('favicon.ico','wb') as f:
    f.write(r.content)
```

```python
#添加请求头
headers = {
    'User-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36'
}
r = requests.get('https://scrape.center',headers=headers)
print(r.text)
```

```python
#session维持
# 为什么维持session:get&post方法可以模拟网页请求，但是相当于不同的session,浏览器打开了不同的页面
# 场景：第一个请求post方法登录了某网站,第二次像获取成功登录后的个人信息，又用get方法请求个人信息页面，两个独立操作，可以通过设置一样的Cookie实现
#但有更简单的方法,利用session对象维护session:模拟登录成功后进行下一步操作时用到
import requests

requests.get("https://www.httpbin.org/cookies/set/number/1234567")
r = requests.get("https://www.httpbin.org/cookies")
print(r.text)#无法成功获取设置的cookie
s = requests.Session()
s.get("https://www.httpbin.org/cookies/set/number/1234567")
r = s.get("https://www.httpbin.org/cookies")
print(r.text)
```
